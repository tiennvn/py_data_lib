```python
result = pd.Series(data=np.random.randint(1, 1001, 1000), 
                   index=pd.date_range('2010/02/09', periods=1000)
                  )
```

```python
plt.plot(pd.date_range('2010/02/09', periods=1000), cum_sum)
```
